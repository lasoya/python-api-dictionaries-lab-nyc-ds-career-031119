### Q1: Load the first page of results with 50 results per page

```python
import requests
import pandas as pd
url = 'https://api.openbrewerydb.org/breweries'
parameters1 = {'page': 1, 'per_page': 50}                   # defining parameters for api call
response1 = requests.get(url, params=parameters1)   # api call to open brewery
data1 = response1.json()                                          # convert api call response to json
# for i in range(len(data)):
#     open_brew = pd.DataFrame(data[i], )
# print(open_brew.keys())

# for i in range(len(response1.json())):
#     print(data1[i])

print(response1.content)
```

### Q2: This is only the first 50 results.  Get the next 50 and put them together.

```python
parameters2 = {'page': 2, 'per_page': 50}                  # defining parameters for api call
response2 = requests.get(url, params=parameters2)   # api call to open brewery
data2 = response2.json()                                          # convert api call response to json
# for i in range(len(data2)):
#     print(data2[i])
print(response2.content)
```

```python
d1 = pd.DataFrame(data1)     # convert json data from api into pandas DataFrame
d2 = pd.DataFrame(data2)     # convert json data from api into pandas DataFrame
d12 = [d1, d2]                      # assign the 2 datasets to be combined to a variable
data = pd.concat(d12)           #combine the first set of 50 results with the second set of 50 results for a dataset of 100
data
```
